# DOEC Ælfric TOC

Generates a JSON dictionary pairing DOEC file references to works by Ælfric with their DOEC short short titles. Assumes DOEC in XML format in `doec_location` (I haven't tried it with the 2000 SGML corpus). I haven't yet added Genesis chapters 1 to 24 verse 22 (`T06200`, `T06210`), and I have probably overlooked another text or two.

```python
import os,json
from lxml import etree
```

```python
parser = etree.XMLParser(remove_blank_text=True)
doec_location = '../DOEC/xml-corpus/'

def generate_toc(doec_location):
    toc = dict()
    with open('doec-aelfric-filerefs.json', 'r') as json_file:
        aelfric_list = json.load(json_file)
    for reference in aelfric_list:
        file = doec_location + reference + '.xml'
        tree = etree.parse(file, parser=parser)
        root = tree.getroot()
        short_title = root.find("teiHeader/fileDesc/titleStmt/title[@type=\"ss\"]")
        toc[reference] = short_title.text
    return toc
```

```python
if not (os.path.isfile('doec-aelfric_toc.json')):
    aelfric_toc = generate_toc(doec_location)
    with open('doec-aelfric-toc.json', 'w', encoding='utf-8') as outfile:
        json.dump(aelfric_toc, outfile, ensure_ascii=False, indent=4)
```
